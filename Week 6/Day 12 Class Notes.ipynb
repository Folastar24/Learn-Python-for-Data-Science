```python
#Class implementation
#None in python means its an empty value
#f string means that its possible that the string will have some varaibles specfied in between. This is way to assert dynamic values in string
#

class Employee:
    emp_list=[
        {
            "employeeNo": 1,
            "name": "Sam Justin",
            "age": 30,
            "gender": "male"
        },
        {"employeeNo": 2,
            "name": "Mary Akin",
            "age": 31,
            "gender": "female"
        },
        {"employeeNo": 3,
            "name": "Uche Okoli",
            "age": 29,
            "gender": "male"
        }


    ]
    def __init__(self,  employeeNo=None):
        self.employeeNo = employeeNo
        

    def search(self, employeeNo):
        self.employeeNo = employeeNo
        name = None
        age = None
        for i in self.emp_list:
            if i["employeeNo"] == self.employeeNo:
               name = i["name"]
               age = i["age"]
               break
        if name and age:
            print("name:", name)
            print("age:", age)
        else:
            print(f"No employee record matching {self.employeeNo} was found")

    def add_record(self, name, age, gender, employeeNo):
        new_record = {
            "employeeNo": employeeNo,
            "name": name,
            "age": age,
            "gender": gender
        }
        self.emp_list.append(new_record)
        return new_record


    def get_all_records(self):
        print(self.emp_list)


emp = Employee()
emp.search(10)

```
